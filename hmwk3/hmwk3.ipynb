```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

train = pd.read_csv("data/train.csv", dtype={"Age": np.float64})
test = pd.read_csv("data/test.csv", dtype={"Age": np.float64})
```

```python
train.info()
```

```python
train.head()
```

```python
# Sex
train.drop(['Ticket', 'Name'], inplace=True, axis=1)
train.Sex.fillna('0', inplace=True)
train.loc[train.Sex != 'male', 'Sex'] = 0
train.loc[train.Sex == 'male', 'Sex'] = 1
# Cabin
train.Cabin.fillna('0', inplace=True)
train.loc[train.Cabin.str[0] == 'A', 'Cabin'] = 1
train.loc[train.Cabin.str[0] == 'B', 'Cabin'] = 2
train.loc[train.Cabin.str[0] == 'C', 'Cabin'] = 3
train.loc[train.Cabin.str[0] == 'D', 'Cabin'] = 4
train.loc[train.Cabin.str[0] == 'E', 'Cabin'] = 5
train.loc[train.Cabin.str[0] == 'F', 'Cabin'] = 6
train.loc[train.Cabin.str[0] == 'G', 'Cabin'] = 7
train.loc[train.Cabin.str[0] == 'T', 'Cabin'] = 8
# Embarked
train.loc[train.Embarked == 'C', 'Embarked'] = 1
train.loc[train.Embarked == 'Q', 'Embarked'] = 2
train.loc[train.Embarked == 'S', 'Embarked'] = 3
train.Embarked.fillna(0, inplace=True)
train.fillna(-1, inplace=True)
```

```python
train.head()
```

```python
train["Child"] = float('NaN')

train.loc[train.Age < 18, "Child"] = 1
train.loc[train.Age >=18, "Child"] = 0
```

```python
train.head()
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
```

```python
train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create a new array with the added features: features_two
features_two = train[["Pclass","Age","Sex","Fare", "SibSp", "Parch", "Embarked"]].values
target = train["Survived"].values
#Control overfitting by setting "max_depth" to 10 and "min_samples_split" to 5 : my_tree_two
max_depth = 10
min_samples_split = 5
my_tree_two = DecisionTreeClassifier(max_depth = 10, min_samples_split = 5, random_state = 1)
my_tree_two = my_tree_two.fit(features_two, target)

#Print the score of the new decison tree
print(my_tree_two.score(features_two, target))
```
